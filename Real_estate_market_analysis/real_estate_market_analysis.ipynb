# Исследование объявлений о продаже квартир


### Описание проекта
В нашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

### Описание данных
- airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
- balcony — число балконов
- ceiling_height — высота потолков (м)
- cityCenters_nearest — расстояние до центра города (м)
- days_exposition — сколько дней было размещено объявление (от публикации до снятия)
- first_day_exposition — дата публикации
- floor — этаж
- floors_total — всего этажей в доме
- is_apartment — апартаменты (булев тип)
- kitchen_area — площадь кухни в квадратных метрах (м²)
- last_price — цена на момент снятия с публикации
- living_area — жилая площадь в квадратных метрах (м²)
- locality_name — название населённого пункта
- open_plan — свободная планировка (булев тип)
- parks_around3000 — число парков в радиусе 3 км
- parks_nearest — расстояние до ближайшего парка (м)
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м)
- rooms — число комнат
- studio — квартира-студия (булев тип)
- total_area — площадь квартиры в квадратных метрах (м²)
- total_images — число фотографий квартиры в объявлении

## Предобработка данных

### Работа с пропусками
Посмотрим на данные в каждом столбце с пропущенными значениями. 

```python
data['ceiling_height'].value_counts() # посмотрим на уникальные значения в столбце ceiling_height
```

```python
data['ceiling_height'].isna().sum() # посчитаем пропуски
```

В данном столбце отсутствие высоты потолков может быть обусловлено тем, что продавец не знает либо не хочет сообщать эту информацию. Возможно продавцы не хотят акцентировать эту информацию, потому что это им невыгодно, плохо повлияет на конечную стоимость объекта. Поскольку замена на медиану или среднее исказит данные по объектам, то лучше оставить эти пропуски.

Также в статистике по таблице отметили, что есть аномальные значения для высоты потолков (например, 100 м). Посчитаем количество таких аномальных значений, взяв за верхнюю границу 4.5 м высоту потолков, так как старый фонд жилья в Санкт-Петербурге имеет высокие потолки.

```python
data[data['ceiling_height'] > 4.5]['ceiling_height'].count()
```

```python
 data['ceiling_height'].median()
```

Заменим эти значения на медианные, потому что они составляют менее 5%. Аномальные значения могут повлиять на результаты анализа.

```python
data.loc[data['ceiling_height'] > 4.5, 'ceiling_height'] = data['ceiling_height'].median()
```

```python
data[data['ceiling_height'] > 4.5]['ceiling_height'].count()
```

```python
data['floors_total'].isna().sum() # посчитаем пропуски в столбце 'floors_total'
```

Пропусков в этом столбце не много. Причиной пропуска может быть как случайность, так и заведомо не указанная информация (например, чтобы покупатели не узнали, что квартира на последнем этаже). Эти данные лучше оставить пустыми, потому что найти подходящее значение на замену не получится, а замена на медианное значение может исказить результат анализа.

```python
data['living_area'].isna().sum() # посчитаем пропуски в столбце'living_area'
```

Пропусков в столбце с жилой площадью около 10%. Причиной пропуска могут быть как отсутствие данных у продавца, случайный пропуск, так и преднамеренное незаполнение столбца. Найдем долю жилой площади от общей для всех объектов, посчитаем среднюю долю и применим её для подсчета жилой площади в полях, где пропуски. 

```python
data['share_living_area'] = data['living_area'] / data['total_area'] # создадим новый столбец с долей жилой площади от всей площади
avg_living_share = data['share_living_area'].mean() # найдем среднее в этом столбце
avg_living_share
```

```python
data['living_area'] = data['living_area'].fillna(data['total_area']*avg_living_share) # заполним все пропуски в столбце'living_area' значением, полученным умножением средней доли на общую площадь 
data['living_area'].isna().sum()
```

```python
data['kitchen_area'].isna().sum() # посчитаем пропуски в столбце'kitchen_area'
```

```python
# посчитаем сколько квартир типа студия и свободной планировки имеют пропуски в площади кухни
data[(data['kitchen_area'].isna()) & ( (data['open_plan'] == True) |(data['studio'] == True))]['open_plan'].count() 
```

Для объектов типа студия и свободная планировка заменим пропуски по площади кухни на 0, потому что для кухни не выделена отдельная площадь.

```python
data.loc[(data['kitchen_area'].isna()) & ( (data['open_plan'] == True) |(data['studio'] == True)), 'kitchen_area'] = 0
data['kitchen_area'].isna().sum()
```

Заменим оставшиеся пропуски в столбце `kitchen_area` по тому же алгоритму, что и с жилой площадью. Только выполним проверку, не превышает ли сумма жилой площади и площади кухни общую площадь квартиры.

```python
data['share_kitchen_area'] = data['kitchen_area'] / data['total_area'] # создадим новый столбец с долей площади кухни от всей площади
avg_kitchen_share = data['share_kitchen_area'].mean() # найдем среднее в этом столбце
print(avg_kitchen_share)
```

```python
data.loc[((data['total_area'] - data['living_area'] - data['total_area']*avg_kitchen_share) < 0) & (data['kitchen_area'].isna()), 'kitchen_area'] = data['total_area'] - data['living_area']
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area']*avg_kitchen_share) # заполним все оставшиеся пропуски в столбце'kitchen_area' значением, полученным умножением средней доли на общую площадь 
```

```python
data['kitchen_area'].isna().sum()
```

Оставим новые столбцы с долями, они нам пригодятся в будущем.

```python
data['is_apartment'].value_counts() # посмотрим, какие уникальные значения в столбце 'is_apartment'
```

```python
data['is_apartment'].isna().sum() # посчитаем пропуски
```

Если человек ничего не указал в этом поле, то скорее всего квартира не относится к типу апартаментов и можно поставить вместо пропусков значение False.

```python
data['is_apartment'] = data['is_apartment'].fillna(False) # заполним все пропуски значением False
data['is_apartment'].isna().sum()
```

Определить населенный пункт можно по расстоянию до центра. Проверим есть ли значения в столбце `city_center_nearest` для пропусков в столбце `locality_name`

В большинстве случаев расстояние до центра известно. Но есть 8 строк, где неизвестно ни расстояние до центра, ни населенный пункт. Посмотрим на эти строки.

```python
data[(data['locality_name'].isna()) & (data['city_center_nearest'].isna())] 
```

В этих строках очень много пропусков и их количество незначительно, так что можно их удалить.

```python
# посмотрим на минимальные значения расстояния до центра
data.pivot_table(index='locality_name', values='city_center_nearest', aggfunc='min').sort_values(by='city_center_nearest')
```

```python
data.loc[(data['locality_name'].isna()) & (data['city_center_nearest'] < 17000), 'locality_name'] = 'Санкт-Петербуг'
data.loc[(data['locality_name'].isna()) & (data['city_center_nearest'] > 17000), 'locality_name'] = 'Другие населенные пункты'
data['locality_name'].isna().sum()
```

```python
data = data.dropna(subset=['locality_name']) # удаляем оставшиеся 8 строк с пропусками
data['locality_name'].isna().sum()
```

```python
# напишем функцию определения этажа квартиры
def floor_cat (row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == floors_total:
        return 'Последний'
    if floor == 1:
        return 'Первый'
    else:
        return 'Другой'    
```

Чтобы изучить параметры квартир создадим функцию, в которую будем передавать параметр, который нас интересует, и функция будет строить гистограмму, диаграмму размаха и статистику для выбранного столбца.

```python
def parametre_desc(parametre, bins=None, range=None, box_lim=None):
    """возвращает статистику, гистограмму и диаграмму размаха для переданного ей параметра"""
    data[parametre].hist(range=range, bins=bins)
    plt.title('Гистограмма распределения по ' + parametre)
    plt.show()
    data.boxplot(parametre)
    plt.ylim(box_lim)
    plt.title('Диаграмма размаха по ' + parametre)
    plt.show()
    return data[parametre].describe()
```

```python
parametre_desc('total_area', range=(0, 400), bins=100, box_lim=[0, 250])
```

Для каждой квартиры есть информация о расстоянии до центра. Выделим квартиры в Санкт-Петербурге. Надо выяснить, какая область входит в центр.

 Создадим столбец с расстоянием до центра в километрах: округлим до целых значений. 

```python
spb_data = good_data.query('locality_name == "Санкт-Петербург"') # Выделим квартиры в Санкт-Петербурге 
spb_data['city_center_nearest_km'] = round(spb_data['city_center_nearest'] / 1000) # Создадим столбец с расстоянием до центра в километрах
spb_data[['locality_name', 'city_center_nearest', 'city_center_nearest_km']].head() # проверим
```
